### **Model Building :**  

- Model building is the process of creating a model to read the data and make predictions or decisions.

- It is build using a set if training data.

#### **Multi Linear Regression :**

- Multi-linear regression is a statistical method that uses multiple independent variables to predict a single dependent ( target or y ) variable.

**Formula :**

- **y = m0 + m1 x1 + m2 x2 + m3 x3 + ... + mn xn + c**  
( default value of m0 = 0 and c = 0 )  
( x1 + x2 + x3 + ... + xn ) - is the independent variable  
(m1 + m2 + m3 + ... + mn ) - is the correlation coefficient  
y - is the dependent variable  
m0 - is the bias  
c -  is the constant

 image.png

```python
# Import LinearRegression

from sklearn.linear_model import LinearRegression
mlr = LinearRegression()
```

```python
# train the model

mlr.fit(x_train , y_train)
```

```python
# find predicted values

y_pred = mlr.predict(x_test)
y_pred
```

```python
# Visualisation of predicted model

plt.scatter(x_train['Hours'],y_train)
plt.plot(x_test['Hours'],y_pred)
plt.scatter(x_train['sHours'],y_train,c = 'green')
plt.plot(x_test['sHours'],y_pred,c = 'green')
plt.title('TV Marketing')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

#### **Model Evaluation :**

- Model evaluation is the process of evaluate the performance of a model on a separate set of data test data.

```python
# import evaluation metrics for regression model

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

**Coefficient of Determination : ( R - Squared or R)**

- R - Squared measures the proportion of the variance in the dependent variable that is predictable from the independent variables.

- It ranges from 0 to 1, with higher values indicating a better fit of the model to the data.

 image.png

```python
# find r_squared value

R2 = r2_score(y_test, y_pred) # r^2 = 1 - np.var(y_test - y_predict) / np.var(y_test)
R2
```

**Adjusted R - Squared : ( Adjusted - R )**

- Adjusted R-squared is a modified version of R - Squared that adjusts for the number of predictors in the model.

- It penalizes the addition of unnecessary predictors.

```python
# find adjusted r_squared value

Adjusted_R2 = 1 - ((1 - R2) * (len(x) - 1)) / (len(x) - x.shape[1] - 1)
Adjusted_R2
```

**Mean Squared Error ( MSE ) :**

- MSE measures the average squared difference between the predicted values and the actual values.

- It penalizes large errors more than smaller ones.

 image.png

```python
# find mean squared error value

MSE = mean_squared_error(y_test,y_pred) # MSE = np.mean(np.square(y_test - y_pred))
MSE
```

**Root Mean Squared Error ( RMSE ) :**

- RMSE is the square root of the MSE.

- It is in the same unit as the dependent variable, making it easier to interpret.

 image.png

```python
# find root mean squared error value

RMSE = np.sqrt(MSE)
RMSE
```

**Mean Absolute Error ( MAE ) :**

- MAE measures the average absolute difference between the predicted values and the actual values.

- It is calculated as the average of the absolute differences between predicted and actual values.

 image.png

```python
# find mean absolute error value

MAE = mean_absolute_error(y_test,y_pred) # MAE = np.mean(np.abs(y_test - y_predict))
MAE
```

**Mean Absolute Percentage Error ( MAPE ) :**

- MAPE measures the average percentage difference between the predicted and actual values.

- It is useful when the scale of the data varies widely.

 image.png

```python
MAPE = mean_absolute_percentage_error(y_test,y_pred)
MAPE
```

- Mean_Squared_Error ( MSE ) = 32.721 suggests that, on average, the squared difference between the predicted and actual values is 32.721. A lower MSE indicates that the model's predictions are closer to the actual values, suggesting better accuracy.
