```python
# Saparate the Indepentend variable

x = df.iloc[:,:-1]
x
```

**Train & Test Split :**

- Split the dataset into train and test sets.

```python
# Split the train & test data

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=1/3,random_state=7)
```

### **Model Building :**  

- Model building is the process of creating a model to read the data and make predictions or decisions.

- It is build using a set if training data.

**Simple Linear Regresion :**

- Simple linear regression is a statistical method uses only one independent variable to predict a single dependent ( target or y ) variable.

**Formula :**

- **y = m x + c**  
( default value of m = 1 and c = 0 )   
y - is the dependent variable  
x - is the independent variable  
m - is the bias  
c - is the constant

 image.png


```python
# Import LinearRegression

from sklearn.linear_model import LinearRegression
slr = LinearRegression()
```

```python
slr.fit(x_train, y_train)
```

```python
y_predict = slr.predict(x_test)
y_predict
```

```python
# Visualisation of prdicted model

plt.scatter(x_train,y_train)
plt.plot(x_test,y_predict,c='red')
plt.title('TV Marketing')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

#### **Model Evaluation :**

- Model evaluation is the process of evaluate the performance of a model on a separate set of data test data.

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
# find r_squared value

r2 = r2_score(y_test, y_predict) # r^2 = 1 - np.var(y_test - y_predict) / np.var(y_test)
r2
```

```python
# find adjusted r_squard value

adjusted_r2 = 1 - ((1 - r2) * (len(x) - 1)) / (len(x) - x.shape[1] - 1)
adjusted_r2
```

```python
# find mean square error value

MSE = mean_squared_error(y_test,y_predict) # MSE = np.mean(np.square(y_test - y_predict))
MSE
```

```python
# find root mean square error value

RMSE = np.sqrt(MSE)
RMSE
```

```python
# find mean absolute error value

MAE = mean_absolute_error(y_test,y_predict) # MAE = np.mean(np.abs(y_test - y_predict))
MAE
```

```python
# find log of root mean square error value

LRMSE  = np.log(RMSE)
LRMSE
```
